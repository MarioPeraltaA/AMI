## Forecasting of Hosting Capacity

### Daily at base case V-P

```python
import dsshc as ckt
DSScircuit = ckt.DSScircuit
df = ckt.run_Daily(DSScircuit)
V_df = df[0]
P_df = df[1]
Q_df = df[2]

loadNames = DSScircuit.Loads.AllNames
```

```python
from scipy import stats
import numpy as np


def Plim(reg, lim_inf=-15, lim_sup=1):
    p = np.linspace(lim_inf, lim_sup, 10000)
    b = reg.intercept
    m = reg.slope

    y = lambda x: m*x + b
    v = list(map(y, p))
    for n, V in enumerate(v):
        if m < 0:
            if 1.0499 < V < 1.05:
                return (V, p[n])
        elif m >= 0:
            if 0.95 < V < 0.9501:
                return (V, p[n])


dailyHC = []
oddload = []
for loadID in loadNames:
    V = V_df[loadID].values / 240
    P = P_df[loadID].values
    if all([v==0 for v in V]) or all([p==0 for p in P]):
        continue
    else:
        reg = stats.linregress(P, V)
        VP = Plim(reg)
        if VP:
            dailyHC.append(VP[1])
        else:
            oddload.append(loadID)

print(len(oddload))
print(len(dailyHC))
```

### Vmax throughout one month

```python
import dsshc as ckt

DSScircuit = ckt.DSScircuit
data = ckt.get_Vmax(DSScircuit)
```

```python
import numpy as np
hc_group = data["Group2"]
vmax_data = []
p_data = []
for load in hc_group.values():
    vmax_data += load["Vmax"]
    p_data += load["Pmax"]

P = np.array(p_data)
V = np.array(vmax_data)
V = V/240
```

```python
from scipy import stats
import matplotlib.pyplot as plt
reg = stats.linregress(P, V)
b = reg.intercept
m = reg.slope
p = np.linspace(-10, 1, 1000)
plt.figure(figsize=(8, 2))
plt.plot(P, V, 'o')
plt.plot(p, m*p + b, "orange")
# plt.plot(-7.71977197719772, 1.0499981076159115, "o")
# plt.plot(-18.055805580558054, 1.049999229042726, "o")
plt.plot(-6.556655665566556, 1.0499956272568078, "o")


plt.title('H.C. ModelBased Grupo2')
plt.xlabel('P[kW]')
plt.ylabel('V[pu]')
plt.legend(["V",
            "linear fit",
            "(-6.56, 1.05)"])
# plt.savefig("regG2.pdf")
plt.show()
```
